```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("results/combined.csv", index_col=0)
```

```python
df.iloc[0]
```

```python
df = df[df["metric"] != "Full_Predictions"]
df["value"] = df["value"].astype(float)
df["value"].iloc[0]
```

```python
broken = df.loc[(df["metric"] == "abag_test_loss") & (df["value"] > 20), "run"]
```

```python
ax = sns.boxplot(data=df[~df["run"].isin(broken)], y="value", x="metric", hue="benchmark_type")
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")
ax.set_ylim([0.0, 0.4])
```
