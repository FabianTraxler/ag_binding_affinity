```python
import wandb
import pandas as pd
import numpy as np
```

```python
api = wandb.Api()
sweep = api.sweep("dachdiffusion/abag_binding_affinity/2tsahoiy")
```

```python
datalist = []
for run in sweep.runs:
    print('run_length', len(run.history()))
    datalist.append(run.history())
    
data = pd.DataFrame(datalist)
print('data length', len(data))
```

```python
data = pd.concat(datalist)
print(len(data))
```

```python
for col in data.columns:
    if 'gradients' not in col and 'parameter' not in col:
    #if 'test' in col or 'loss' in col or 'pearson' in col or 'corr' in col or 'rmse' in col
        print(col)
```

```python
benchmark_data = data[~data['skempi_test_pearson_grouped_mean'].isna()]
```

```python
benchmark_data[~benchmark_data['abag_affinity#L1-1+relative_cdf-1_val_corr'].isna()][['abag_affinity#L1-1+relative_cdf-1_val_corr', 'val_loss']]
```

```python
# general correlation

#abag_train_pearson
#train_loss
#abag_test_loss
#benchmark_test_loss
#skempi_test_pearson_grouped_mean
#val_loss
#abag_train_loss
#pearson_correlation
#abag_test_pearson
#benchmark_test_pearson
#skempi_test_pearson
#skempi_test_loss

comp_losses = ['abag_train_pearson', 'pearson_correlation']
benchmark_losses = ['skempi_test_pearson_grouped_mean', 'benchmark_test_pearson', 'abag_test_pearson']

print("General correlation between pearson measures across all runs:")
for comp_loss in comp_losses:
    for bench_loss in benchmark_losses:
        sub_data = benchmark_data[~benchmark_data[comp_loss].isna()]
        sub_data = sub_data[~sub_data[bench_loss].isna()]
        print(f'\tcorrelation between {comp_loss} and {bench_loss} is {np.corrcoef(sub_data[comp_loss], sub_data[bench_loss])[0, 1]}')
```

```python
loss_criterions = ['abag_affinity#L1-1+relative_ce-1_val_acc',
'abag_affinity#L1-1+relative_ce-1_val_loss',
'abag_affinity#L1-1+relative_ce-1_val_rmse',
'abag_affinity#L1-1+relative_cdf-1_val_corr',
'abag_affinity#L1-1+relative_cdf-1_val_rmse',
'abag_affinity#L1-1+relative_cdf-1_val_acc',
'abag_affinity#L1-1+relative_cdf-1_val_loss',
'abag_affinity#L1-1+relative_L1-1_val_corr',
'abag_affinity#L1-1+relative_L1-1_val_loss',
'abag_affinity#L1-1+relative_L1-1_val_rmse',
'abag_affinity#L1-1+relative_L1-1_val_acc',
'abag_affinity#L1-1_val_loss',
'abag_affinity#L1-1_val_rmse',
'abag_affinity#L1-1_val_corr',]
benchmark_losses = ['skempi_test_loss', 'benchmark_test_loss', 'abag_test_loss']
print("Correlation between the different loss criteria and the benchmark losses:")
for loss_crit in loss_criterions:
    print('')
    print('\tCorrelations for ', loss_crit)
    for bench_loss in benchmark_losses:
        sub_data = benchmark_data[~benchmark_data[loss_crit].isna()]
        sub_data = sub_data[~sub_data[bench_loss].isna()]
        print(f'\t\tcorrelation with {bench_loss} is {np.corrcoef(sub_data[loss_crit], sub_data[bench_loss])[0, 1]}')
```
